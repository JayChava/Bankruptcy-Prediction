# Importing the required Libraries

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

# Importing comp3 File

```python
comp = pd.read_sas('comp3.sas7bdat')
comp.head()
```

```python
comp.columns = comp.columns.str.lower() #Code to convert dataframe columns to lower case
comp['cik']=comp['cik'].astype(str)     # Conveting cik column to string type
comp['cik']=comp['cik'].astype(float)   # Converting cik column to float type
comp['cik']=comp['cik'].astype(int)     # Converting cik column to int type
comp['sic'] = comp['sic'].apply(lambda x : int(x))  # Applying lambda function to convert every data point to integer
```

# Reading Bankruptcy File

```python
bankruptcy = pd.read_sas('Bankruptcy_20002019.sas7bdat')
bankruptcy.head()
```

```python
bankruptcy.columns = bankruptcy.columns.str.lower() #Converting bankruptcy columns to lower case
bytes_string_list = []    # Initializing a list
for elem in bankruptcy['company_fkey']: 
    bytes_string_list.append(elem.decode('utf-8'))    #Decoding every element belonging to company_fkey column
bankruptcy['company_fkey'] = bytes_string_list         
bankruptcy['company_fkey'] = bankruptcy['company_fkey'].apply(lambda x: x.lstrip('0'))  # Stripping 0 from the left
bankruptcy['company_fkey'] = bankruptcy['company_fkey'].astype(int)            #Converting company_fkey to int type
bankruptcy.rename(columns={'company_fkey':'cik'},inplace=True)               # renaming the column company_fkey to cik
bankruptcy.head()
```

# Code for checking which cik's are not there in bankruptcy and need their bankruptcy to be indicated as 0

```python
comps_cik = comp['cik'].values.tolist()
bank_cik = bankruptcy['cik'].values.tolist()
ciks_zero = []
for cik in comps_cik:
    if cik not in bank_cik:
        ciks_zero.append(cik)
len(ciks_zero)
```

```python
zero_bank = pd.DataFrame()
distinct_ciks = list(set(ciks_zero))
zero_comp = comp[comp['cik'].isin(distinct_ciks)]
```

```python
ss = [0 for i in range(len(zero_comp))]
zero_comp['bank_list'] = ss
```

# Proceeding to merge dataframes (comp3, bakruptcy) in order to indicate those firms as 1 which have been bankrupted within one year of filing and 0 for those companies which were filed for bankruptcy but didn't get bankrupt within one year of filing

```python
merged = pd.merge(comp,bankruptcy,on='cik',how='inner')
```

```python
len(merged)
```

```python
merged['filingdate'] = merged['filing'].astype(str)
merged['bankdate'] = merged['bank_begin_date'].astype(str)
merged['results'] = merged['filingdate'] > merged['bankdate']
```

```python
merged['results'].value_counts()
```

```python
def rowIndex(row):
    return row.name
```

```python
merged['day_diff'] = (merged['bank_begin_date'] - merged['filing']).dt.days
merged['index'] = merged.apply(rowIndex, axis=1)
bank_list = []
for elem in merged['day_diff']:
    if elem < 365 and elem >0:
        bank_list.append(1)
    else:
        bank_list.append(0)
merged['bank_list'] = bank_list
narrow = merged[['cik','filing','bank_begin_date','day_diff','index','bank_list']]
narrow.head()
```

# Declaring dictionaries to keep a track of those cik's which have multiple bankruptcy dates and single bankruptcy date

```python
dup = bankruptcy[bankruptcy.cik.duplicated(keep=False)]['cik']
count_dict = {}
for elem in dup:
    if elem in count_dict:
        count_dict[elem] += 1
    else:
        count_dict[elem] = 1
bankcik = bankruptcy['cik'].values.tolist()
bcik_dict = {}
for elem in bankcik:
    if elem in bcik_dict:
        bcik_dict[elem] += 1
    else:
        bcik_dict[elem] = 1
count_one = 0
count_two = 0
for k,v in bcik_dict.items():
    if v == 1:
        count_one += 1
    else:
        count_two += 1
print(count_one)
print(count_two)
```

# Filtering for those cik's which have multiple bankruptcy dates

```python
duplicate_keys= []
for k,v in count_dict.items():
    if v == 2:
        duplicate_keys.append(k)
```

```python
bank_dupl = pd.DataFrame()
for elem in duplicate_keys:
    bank_dupl = bank_dupl.append(merged[merged['cik']==elem])
```

## Code for removing duplicates

```python
narrowlist = bank_dupl[['cik','filing','bank_begin_date','day_diff','index','bank_list']].values.tolist()
wanted_list= []
for i in range(len(narrowlist)):
    try:
        if narrowlist[i][5] == 1:
            wanted_list.append(narrowlist[i][4])
        elif (narrowlist[i][1] == narrowlist[i+1][1]):
            if narrowlist[i][3] > 0:
                if narrowlist[i][3] < narrowlist[i+1][3]:
                    wanted_list.append(narrowlist[i][4])
            elif narrowlist[i][3] < 0:
                wanted_list.append(narrowlist[i+1][4])
    except IndexError:
        print('oob')
wanted_list = list(set(wanted_list))
dupl_filtered = bank_dupl[bank_dupl.index.isin(wanted_list)]
```

```python
dupl_filtered[['cik','filing','bank_begin_date','day_diff','index','bank_list','results']].head()
```

```python
len(dupl_filtered)
```

# Code for assigning to unique cik's

```python
unique_keys= []
for k,v in bcik_dict.items():
    if v == 1:
        unique_keys.append(k)
```

```python
final_bank = pd.DataFrame()
final_bank = merged[merged['cik'].isin(unique_keys)]
final_bank.head()
```

```python
narrowlist = final_bank[['cik','filing','bank_begin_date','day_diff','index','bank_list']].values.tolist()
wanted_list= []
for i in range(len(narrowlist)):
    try:
        if narrowlist[i][5] == 1:
            wanted_list.append(narrowlist[i][4])
        elif (narrowlist[i][1] == narrowlist[i+1][1]):
            if narrowlist[i][3] > 0:
                if narrowlist[i][3] < narrowlist[i+1][3]:
                    wanted_list.append(narrowlist[i][4])
            elif narrowlist[i][3] < 0:
                wanted_list.append(narrowlist[i+1][4])
        elif narrowlist[i][1] != narrowlist[i+1][1]:
            wanted_list.append(narrowlist[i][4])
    except IndexError:
        print('oob')
wanted_list = list(set(wanted_list))
remove_dupl = final_bank[final_bank.index.isin(wanted_list)]
remove_dupl.head()
```
